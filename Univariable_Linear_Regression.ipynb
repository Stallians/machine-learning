```python
global learning_rate
learning_rate=0.1

#equation of line h(x) = mx+b
def linehx(m,x,b): 
    return m*x+b

#cost function J(m,b)
def costfn(m,x,b,y):
    error=0
    for xval,yval in zip(x,y):
        error+=(linehx(m,xval,b)-yval)**2
    return 0.5*N*error

#partial derivative w.r.t b
def pder_b(m,x,b,y):
    err=0
    for xval,yval in zip(x,y):
        err+=(m*xval+b-yval)
    return err/N

#partial derivative w.r.t m
def pder_m(m,x,b,y):
    err=0
    for xval,yval in zip(x,y):
        err+=(m*xval+b-yval)*xval
    return err/N

# N = number of records or rows
def declareN(value):
    global N
    N=value  

# new values of 'm' and 'b' using gradient descent
def grad_desc(m,x,b,y):
    tempm=m-learning_rate*pder_m(m,x,b,y)
    tempb=b-learning_rate*pder_b(m,x,b,y)
    return tempm,tempb

# minimising cost using values from gradient descent iteratively
def fitt(m,x,b,y):
    count=1
    c=costfn(m,x,b,y)
    corr=grad_desc(m,x,b,y)
    print "[ cost=%f\tnew_m=%f\tnew_b=%f ]"%(c,corr[0],corr[1])
    c_new=costfn(corr[0],x,corr[1],y)
    while c_new<c:
        count+=1
        c=c_new
        corr=grad_desc(corr[0],x,corr[1],y)
        print "[ cost=%f\tnew_m=%f\tnew_b=%f ]"%(c,corr[0],corr[1])
        c_new=costfn(corr[0],x,corr[1],y)
    print "Final: m=%f\tb=%f\tCount=%d"%(corr[0],corr[1],count)
```
